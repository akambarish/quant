```python
import plotly.express as px
px.line(data)
```

```python
returns= data.pct_change()
returns
```

```python
px.line(returns)
```

```python
weighted_model_portfolio = returns.mul(weights, axis=1)
weighted_model_portfolio
```

```python
returns['Portfolio'] = weighted_model_portfolio.sum(axis=1).dropna()
returns['Portfolio'] = returns['Portfolio'] * 100
returns['Portfolio']
```

```python
var95 = np.percentile(returns['Portfolio'], 5)
var95
```
